# Import Libraries

```python
import community.community_louvain as community
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import pandas as pd
import numpy as np

import os, sys
nb_dir = os.path.split(os.getcwd())[0]
if nb_dir not in sys.path:
    sys.path.append(nb_dir)
    
from Utils.utils import shannon_entropy, extract_backbones_and_measures, extract_measures
from Backbones.high_salience_skeleton import high_salience_skeleton as hss
from Backbones.doubly_stochastic import read, doubly_stochastic as ds
from Backbones.marginal_likelihood_filter import MLF
from Backbones import disparity_filter as disf
from Backbones import h_backbone as hb
#from Backbones import marginal_likelihood_filter as ml
from Backbones.noise_corrected import noise_corrected as nc
%matplotlib inline
```

# Import the network

```python
#initialize the network name 
network= 'lesmis'

#read edge list from csv file
edge_list = pd.read_csv('../Datasets/' + network + '.csv')

#read edge list from csv file for the doubly stochastic filter, noice corrected and the high salience skeleton
table, nnodes, nnedges = read("../Datasets/" + network + '.csv', "weight", sep=',', consider_self_loops=False, triangular_input = True, undirected=True) 
    
#create graph from the edge list
G = nx.from_pandas_edgelist(edge_list, edge_attr='weight', create_using=nx.Graph())
    
#take only the largest connected component
largest_cc = max(nx.connected_components(G), key=len)
G = G.subgraph(largest_cc).copy()
        
original_measures = extract_measures(G)
```

# Extract the backbones

```python
#initialize the fraction of nodes to be extracted
fractions = [0.02, 0.04, 0.06, 0.08, 0.1 , 0.12, 0.14, 0.16, 0.18, 0.2 , 0.22, 0.24, 0.26, 0.28, 0.3 , 0.32, 0.34, 0.36, 0.38, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9]

#initialize the dictionary that contains all backbones
all_backbones_graphs = dict()
all_backbones_measures = dict()
```

## Naive Threshold

```python
#sort edges by weight
backbone = edge_list.copy()
backbone.sort_values(by='weight', ascending=False)


#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, backbone, fractions, 'Edges')

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['Naive Threshold'] = backbones_dict
all_backbones_measures['Naive Threshold'] = backbone_measures
```

## The Disparity Filter Backbone

```python
#apply the disparity filter algorithm
backbone = disf.disparity_filter(G)

#convert the graph to a dataframe for processing and sort the values by the alpha value
df_backbone = nx.to_pandas_edgelist(backbone)
df_backbone = df_backbone.rename(columns={'alpha': 'threshold'})
df_backbone = df_backbone.sort_values(by='threshold')

#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, df_backbone, fractions, 'Edges')

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['Disparity Filter'] = backbones_dict
all_backbones_measures['Disparity Filter'] = backbone_measures
```

## The High Saleince Skeleton

```python
#apply the high salience skeleton algorithm
hss_backbone = hss(table, return_self_loops=False, undirected=True)
    
#sort score values of the edges
hss_backbone = hss_backbone.rename(columns={'score': 'threshold'})
hss_backbone = hss_backbone.sort_values(by='threshold', ascending=False)

#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, hss_backbone, fractions, 'Edges')

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['High Salience Skeleton'] = backbones_dict
all_backbones_measures['High Salience Skeleton'] = backbone_measures
```

## The Doubly Stochastic Filter

```python
#apply the doubly stochastic filter
ds_backbone = ds(table, undirected = True, return_self_loops = False)

#sort score values of the edges
ds_backbone = ds_backbone.rename(columns={'score': 'threshold'})
ds_backbone = ds_backbone.sort_values(by='threshold', ascending=False)
    

#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, ds_backbone, fractions, 'Edges')

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['Doubly Stochastic Filter'] = backbones_dict
all_backbones_measures['Doubly Stochastic Filter'] = backbone_measures
```

## The Marginal Likelihood Filter

```python
# Apply the transformer to the edgelist
mlf = MLF(directed=False)
mlf_backbone = mlf.fit_transform(edge_list)

#sort score values of the edges
mlf_backbone = mlf_backbone.rename(columns={'significance': 'threshold'})
mlf_backbone = mlf_backbone.sort_values(by='threshold', ascending=False)

#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, mlf_backbone, fractions, 'Edges')

#fix thresholds since they are transformed to log values in the extraction function
backbone_measures['threshold'] = [np.exp(-threshold) for threshold in backbone_measures['threshold']]

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['Marginal Likelihood Filter'] = backbones_dict
all_backbones_measures['Marginal Likelihood Filter'] = backbone_measures
```

## The Noise-Corrected Filter

```python
#apply the noise corrected filter
nc_backbone = nc(table, undirected = True, return_self_loops = False, calculate_p_value = False)
nc_backbone['threshold'] = nc_backbone["score"] - (1 * nc_backbone["sdev_cij"])

#sort score values of the edges and take only positive values
nc_backbone = nc_backbone.sort_values(by='threshold')
nc_backbone = nc_backbone[nc_backbone['threshold'] >0]

#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, nc_backbone, fractions, 'Edges')

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['Noise Corrected'] = backbones_dict
all_backbones_measures['Noise Corrected'] = backbone_measures
```

## The Polya Filter

```python
#read edge list from csv file
pf_backbone = pd.read_csv('../Results/Backbones Results/PF/' + network + '.csv')

#sort score values of the edges
pf_backbone = pf_backbone.rename(columns={'p_values': 'threshold'})
pf_backbone = pf_backbone.sort_values(by='threshold')

#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, pf_backbone, fractions, 'Edges')

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['Polya Filter'] = backbones_dict
all_backbones_measures['Polya Filter'] = backbone_measures
```

## Gloss

```python
#read edge list from csv file
gloss_backbone = pd.read_csv('../Results/Backbones Results/GLOSS/' + network + '.csv', names=['source', 'target', 'threshold', 'weight'], sep=' ')
labels = pd.read_csv('../Datasets/' + network + '_labels.csv', index_col='index')

#relabel using saved labels
labels = labels.T.to_dict('list')
gloss_backbone['source'] = gloss_backbone['source'].replace(labels)
gloss_backbone['target'] = gloss_backbone['target'].replace(labels)

#sort score values of the edges
gloss_backbone = gloss_backbone.sort_values(by='threshold')

#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, gloss_backbone, fractions, 'Edges')

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['GLOSS Filter'] = backbones_dict
all_backbones_measures['GLOSS Filter'] = backbone_measures
```

## Lans

```python
#read edge list from csv file
lans_backbone = pd.read_csv('../Results/Backbones Results/LANS/' + network + '.csv')

labels = pd.read_csv('../Datasets/' + network + '_labels.csv', index_col='index')

#relabel using saved labels
labels = labels.T.to_dict('list')
lans_backbone['source'] = lans_backbone['source'].replace(labels)
lans_backbone['target'] = lans_backbone['target'].replace(labels)

#sort score values of the edges
lans_backbone = lans_backbone.rename(columns={'weight': 'threshold'})
lans_backbone = lans_backbone.sort_values(by='threshold')

#add weights
weights = []
for i in range(lans_backbone.shape[0]):
    u = lans_backbone.loc[i]['source']
    v = lans_backbone.loc[i]['target']
    weights.append(G[u][v]['weight'])
lans_backbone['weight'] = weights


#extract different backbones with different network size and calculate the evaluation measures
backbones_dict, backbone_measures = extract_backbones_and_measures(G, lans_backbone, fractions, 'Edges')

#add the backbone graphs and measures to the all backbones dictionary
all_backbones_graphs['LANS Filter'] = backbones_dict
all_backbones_measures['LANS Filter'] = backbone_measures
```

# Plot and compare the measures

```python
sns.set_theme()
fig, axs = plt.subplots(5, 2, figsize=(10,20))
for backbone in all_backbones_measures:
    
    axs[0, 0].plot(fractions, all_backbones_measures[backbone]['nodes_fraction'], label=backbone, marker='o', markersize=5)
    axs[0, 0].set_ylabel('Node Fraction')
    
    axs[0, 1].plot(fractions, all_backbones_measures[backbone]['edge_fraction'], label=backbone, marker='o', markersize=5)
    axs[0, 1].set_ylabel('Edge Fraction')
    
    axs[1, 0].plot(fractions, all_backbones_measures[backbone]['nb_connected_components'], label=backbone, marker='o', markersize=5)
    axs[1, 0].set_ylabel('nb_connected_components')
    
    axs[1, 1].plot(fractions, all_backbones_measures[backbone]['average_weighted_degree'], label=backbone, marker='o', markersize=5)
    axs[1, 1].set_ylabel('Average Weighted Degree')

    axs[2, 0].plot(fractions, all_backbones_measures[backbone]['average_link_weight'], label=backbone, marker='o', markersize=5)
    axs[2, 0].set_ylabel('Average Link Weight')
    
    axs[2, 1].plot(fractions, all_backbones_measures[backbone]['average_betweeness'], label=backbone, marker='o', markersize=5)
    axs[2, 1].set_ylabel('Average Betweeness')

    axs[3, 0].plot(fractions, all_backbones_measures[backbone]['weighted_modularity'], label=backbone, marker='o', markersize=5)
    axs[3, 0].set_ylabel('Weighted Modularity')

    axs[3, 1].plot(fractions, all_backbones_measures[backbone]['entropy'], label=backbone, marker='o', markersize=5)
    axs[3, 1].set_ylabel('Entropy')

    axs[4, 0].plot(fractions, all_backbones_measures[backbone]['density'], label=backbone, marker='o', markersize=5)
    axs[4, 0].set_ylabel('Density')
    
    axs[4, 1].plot(fractions, all_backbones_measures[backbone]['diameter'], label=backbone, marker='o', markersize=5)
    axs[4, 1].set_ylabel('Diameter')

#     axs[3, 1].plot(fractions, all_backbones_measures[backbone]['threshold'], label=backbone)
#     axs[3, 1].set_ylabel('Threshold')

axs[0, 0].axhline(y=original_measures['nodes_fraction'], color='black', linestyle='--', label='Original Network')
axs[0, 1].axhline(y=original_measures['edge_fraction'], color='black', linestyle='--', label='Original Network')
axs[1, 0].axhline(y=original_measures['nb_connected_components'], color='black', linestyle='--', label='Original Network')
axs[1, 1].axhline(y=original_measures['average_weighted_degree'], color='black', linestyle='--', label='Original Network')
axs[2, 0].axhline(y=original_measures['average_link_weight'], color='black', linestyle='--', label='Original Network')
axs[2, 1].axhline(y=original_measures['average_betweeness'], color='black', linestyle='--', label='Original Network')
axs[3, 0].axhline(y=original_measures['weighted_modularity'], color='black', linestyle='--', label='Original Network')
axs[3, 1].axhline(y=original_measures['entropy'], color='black', linestyle='--', label='Original Network')
axs[4, 0].axhline(y=original_measures['density'], color='black', linestyle='--', label='Original Network')
axs[4, 1].axhline(y=original_measures['diameter'], color='black', linestyle='--', label='Original Network')



plt.setp(axs[-1, :], xlabel='Fraction of Nodes Preserved')
fig.suptitle('Evaluation Measures for the Les Misrables Network', fontsize=16);

for ax in axs.reshape(-1):
    ax.grid(True)
    
    
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
```

```python
sns.reset_orig()

pos = nx.circular_layout(G)
#pos = nx.shell_layout(G)
fig, axs = plt.subplots(3, 3, figsize=(15,15))

for i, backbone in enumerate(all_backbones_graphs):
    b = all_backbones_graphs[backbone][0.4]
    filtered_edges = [edge for edge in G.edges() if edge not in b.edges()]
    filtered_edges_width = [round((G[u][v]['weight']*len(G.nodes()))/edge_list['weight'].sum(),3) for u,v in filtered_edges]
    backbone_edges_width = [round((G[u][v]['weight']*len(G.nodes()))/edge_list['weight'].sum(),3) for u,v in b.edges()]

    nx.draw_networkx_nodes(G, pos, ax=axs[int(i/3)][i%3], node_color='black', node_size=50)
    nx.draw_networkx_edges(G, pos, edgelist=filtered_edges, width= filtered_edges_width,ax=axs[int(i/3)][i%3], edge_color='silver')
    nx.draw_networkx_edges(G, pos, edgelist=b.edges(), ax=axs[int(i/3)][i%3], edge_color='crimson', width=backbone_edges_width)
    nx.draw_networkx_labels(G, pos, font_size=8, ax=axs[int(i/3)][i%3], font_family="sans-serif", horizontalalignment='center', verticalalignment='bottom')




    axs[int(i/3)][i%3].set_title(backbone, fontsize=18)  

plt.tight_layout()
plt.show()
```
