# Create Graph Using Networkx

```python
G = nx.from_pandas_edgelist(edges_list, edge_attr='weight', create_using=nx.Graph())
```

```python
plt.rcParams['figure.figsize'] = [10, 10]

original = G

pos = nx.spring_layout(original, seed = 1)  # positions for all nodes - seed for reproducibility

# nodes
nx.draw_networkx_nodes(original, pos, node_color='pink', node_size=1750)

# edges
nx.draw_networkx_edges(original, pos, edgelist=original.edges(), width=[d["weight"] for (u, v, d) in original.edges(data=True)])

# labels
nx.draw_networkx_labels(original, pos, font_size=12, font_family="sans-serif", )

ax = plt.gca()
ax.margins(0.08)
plt.title('Original Network', size = 22)
plt.tight_layout()
plt.show()
```

```python

alpha = 0.05
Df_B = disparity_filter(G)
Df_B_list = nx.to_pandas_edgelist(Df_B)
Df_B_graph = nx.from_pandas_edgelist(Df_B_list, edge_attr='weight', create_using=nx.Graph())

#G2 = nx.Graph([(u, v, d) for u, v, d in G.edges(data=True) if d['alpha'] < alpha])
```

```python
backbone = Df_B_graph

significant_links = [(row[0],row[1]) for index, row in Df_B_list.iterrows() if row[3] < 0.05]
non_significant_links = [(row[0],row[1]) for index, row in Df_B_list.iterrows() if row[3]>=0.05]

nx.draw_networkx_nodes(original, pos, node_color='pink', node_size=750)

# edges
nx.draw_networkx_edges(backbone, pos, edgelist=non_significant_links, width=3, edge_color="black", label='Not Significant')
nx.draw_networkx_edges(backbone, pos, edgelist=significant_links, width=3, edge_color="red", label='Significant')

# labels
nx.draw_networkx_labels(original, pos, font_size=12, font_family="sans-serif", )

ax = plt.gca()
ax.margins(0.08)
ax.legend()
plt.title('The Disparity Filter Backbone', size = 22)
plt.tight_layout()
plt.show()
```
