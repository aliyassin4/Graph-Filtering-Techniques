```python
# initialize the network name and dataset path
network = 'airports'
path = '../Airports/'

# read edge list from csv file
edge_list = pd.read_csv( path+ network + '.csv')
edge_list.weight = edge_list.weight.astype(int)
```

```python
labels = pd.read_csv('airports_labels.csv')
north_america = pd.read_csv('../Results/All Backbones/north america/countries.csv')
```

```python
north_america_countries = north_america.drop(columns=['City']).dropna()
north_america_cities = north_america.drop(columns=['Country']).dropna()
df_north_america_airports = labels.merge(north_america_countries, on='Country', how='inner')
df_north_america_airports = df_north_america_airports.append(labels.merge(north_america_cities, on='City'), ignore_index=True) 
df_north_america_airports.sort_values(by='id')
```

```python
df_north_america_edges = edge_list.loc[(edge_list['source'].isin(df_north_america_airports.id)) & (edge_list['target'].isin(df_north_america_airports.id))]
df_north_america_edges = df_north_america_edges.reset_index(drop=True)
df_north_america_edges
```

```python
G = nx.from_pandas_edgelist(df_north_america_edges, edge_attr='weight')
G = ev.get_lcc(G)
df_north_america_lcc = nx.to_pandas_edgelist(G)
```

```python
df_north_america_airports.to_csv('../Results/All Backbones/north america/north_america_nodes.csv', index=False)
df_north_america_edges.to_csv('../Results/All Backbones/north america/north_america_all_components.csv', index=False)
df_north_america_lcc.to_csv('../Results/All Backbones/north america/north_america.csv', index=False)
```
